#21 Why is NumPy’s array slicing faster than Python’s list slicing?
-> NumPy’s slicing is faster than Python’s list slicing because NumPy arrays are contiguous in memory, allow efficient views rather than copies, are homogenous, and are optimized at a low level using C and other performance-enhancing techniques. Python lists, being more general-purpose, carry overhead related to their flexibility and dynamic nature, making slicing slower in comparison.

```python
#2 Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Score': [85, 90, 95, 88]
}

df = pd.DataFrame(data)

# Calculate the mean of the 'Score' column
mean_score = df['Score'].mean()

# Print the result
print("Mean of the 'Score' column:", mean_score)
```

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Math': [90, 80, 70, 60, 85],
    'Science': [85, 78, 72, 65, 80],
    'English': [88, 75, 70, 60, 83],
    'History': [75, 70, 68, 55, 78]
}

df = pd.DataFrame(data)

# Step 1: Calculate the correlation matrix
corr_matrix = df.corr()

# Step 2: Create a heatmap
plt.figure(figsize=(8, 6))  # Optional: set figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#6 Create a DataFrame and add a new column based on an existing column
import pandas as pd

# Step 1: Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 92, 78, 90]
}

df = pd.DataFrame(data)

# Step 2: Add a new column 'Grade' based on the 'Score'
def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'D'

df['Grade'] = df['Score'].apply(assign_grade)

# Step 3: Display the updated DataFrame
print(df)
```

```python
#8 Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]     # Line 1
y2 = [1, 3, 5, 7, 9]      # Line 2
y3 = [5, 3, 4, 2, 1]      # Line 3

# Create the plot
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='green', marker='s')
plt.plot(x, y3, label='Line 3', color='red', marker='^')

# Add title and axis labels
plt.title('Multiple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#9 Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd

# Step 1: Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 92, 78, 90]
}

df = pd.DataFrame(data)

# Step 2: Set a threshold value
threshold = 85

# Step 3: Filter rows where 'Score' is greater than the threshold
filtered_df = df[df['Score'] > threshold]

# Step 4: Display the filtered DataFrame
print(filtered_df)
```

```python
#11  Perform matrix multiplication using NumPy
import numpy as np

# Define two matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Perform matrix multiplication using np.dot()
result = np.dot(matrix1, matrix2)

# Alternatively, you can use the @ operator for matrix multiplication
# result = matrix1 @ matrix2

# Display the result
print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

print("\nResult of Matrix Multiplication:")
print(result)
```
